```python
choice = int(input('Enter the choice'))
if choice ==1:
    print('Ordered item is coffee')
    which_coffee = float(input('Which coffee ?'))
    if which_coffee == 1.1:
        print('Ordered coffee is Latte')
    elif which_coffee ==1.2:
        print('Ordered coffee is Capachino')
    else:
        print('Ordered coffee not available')
elif choice ==2:
    print('Ordered item is tea')
    which_tea = float(input('Which Tea ?'))
    if which_tea == 2.1:
        print('Ordered Tea is Normal Tea')
    elif which_tea ==2.2:
        print('Ordered Tea is Lemon Tea')
    else:
        print('Ordered Tea not available')
elif choice==3:
    print('Ordered item is juice')
    which_juice = float(input('Which Juice ?'))
    if which_juice == 3.1:
        print('Ordered Juice is Apple juice')
    elif which_juice ==3.2:
        print('Ordered Juice is Banana juice')
    else:
        print('Ordered Juice not available')
else:
    print('Invalid choice')
```
